```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
data = datasets.load_breast_cancer()
```

```python
df = pd.DataFrame(data.data[100:110])
df_target = pd.DataFrame(data.target[100:110])
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, df_target, test_size=0.2)
X_test
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
```

```python
a = np.array_split(list(range(data.data.shape[0])), 5)
cv_list = [list(i) for i in a]
```

```python
df = pd.DataFrame(data.data)
df_target = pd.DataFrame(data.target)
```

```python
for i, item in enumerate(cv_list):
    X_test = np.array(df[df.index.isin(cv_list[i])])
    y_test = np.array(df_target[df_target.index.isin(cv_list[i])])

    X_train = np.array(df[df.index.isin(cv_list[i])==False])
    y_train = np.array(df_target[df_target.index.isin(cv_list[i])==False])
    dt.fit(X_train, y_train)
    y_pred = dt.predict(X_test)
    y_test.flatten()
    y_test = y_test.flatten()

    counter = 0
    for i in range(len(y_pred)):
        if y_test[i] == y_pred[i]:
            counter = counter + 1
    print('Accuracy: ', round(counter/len(y_pred),3))
```
